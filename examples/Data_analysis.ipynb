```python
# Den här funktionen skapar listor av alla floatlists (med ground state probability)
#bara för att kunna jobba med filerna

def tofloatlists(dataframe, string): #dataframe is the csv-file that's been read, string is "CVaR cost" or "Ground state probability"
    probs=dataframe[string] 
    floatlists = []
    for i in range(1000):
        probabs = probs[i]
        stringlist=(probabs.lstrip("[").rstrip("]").replace(',', '')).split( )
        floatlist= [float(g) for g in np.array(stringlist)]
        floatlists.append(floatlist)
        
    return floatlists
```

```python
#ett medelvärde av antal iterationer det tar innan COBYLA slutar optimera (samt medelvärde av slutvärdena)

def average_final_value(problists):
    x_lst = []
    y_lst = []
    
    for prob_vector in problists: #bygger en lista med alla slutgiltliga värden på antal iterationer
        x_lst.append(len(prob_vector)) #längden av alla sublistor = antal iterationer innan stopp
        y_lst.append(prob_vector[-1]) #lägger in sista värdet av ground state probability 
        
    meanvalue_x = np.mean(x_lst)
    xerr = np.std(x_lst)
    
    meanvalue_y = np.mean(y_lst) 
    yerr = np.std(y_lst)
    
    return [meanvalue_x,meanvalue_y], [xerr,yerr] 
```

```python
#Denna koden beräknar ett medelvärde av groundstate probability för alla 1000 mätningar för varje iteration.
#I de fall då den slutar optimera tidigre används det sista värdet i resterande medelvärdesbildningar
#kan användas för CVaR cost också

def average_value_prob(problists): #inputen ska vara en problists (alltså en lista av ground state probabilities för en mätserie) 
    average_value_list = []
    yerr = [] 
    
    len_max_problist = len(max(problists, key=len)) #returnerar den längsta sublisten i problists -- alla måste bli så långa
    
    #måste appenda värden i slutet av listan (som är samma som sista värdet) så att alla listor blir lika långa som sista
    for sublist in problists:
        while len(sublist) < len_max_problist:
            sublist.append(sublist[-1])
    
    for i in range(len_max_problist): #beräknar medelvärdet och variansen (för error-bars) för varje iteration
        lst = [item[i] for item in problists]
        average_value_list.append(np.mean(lst)) #medelvärdet 
        yerr.append(np.std(lst)/10) #standradavvikelse (varians) OBS:dividerad med 10 för att få plats i bilden
        
    return average_value_list, yerr
```

```python
# Koden som plottar medelvärdet av alla 1000 mätningar för olika alpha

average_value_list005, yerr005 = average_value_prob(problist_alpha005)
average_value_list010, yerr010 = average_value_prob(problist_alpha010)
average_value_list025, yerr025 = average_value_prob(problist_alpha025)
average_value_list100, yerr100 = average_value_prob(problist_alpha100)


fig, ax = plt.subplots()

for i in range(len(average_value_list005)):
    ax.plot(i+1, average_value_list005[i], 'r.', linewidth=1)
    
for i in range(len(average_value_list010)):
    ax.plot(i+1, average_value_list010[i], 'yx', linewidth=1)
    
for i in range(len(average_value_list025)):
    ax.plot(i+1, average_value_list025[i], 'b+', linewidth=1)
    
for i in range(len(average_value_list100)):
    ax.plot(i+1, average_value_list100[i], 'm1', linewidth=1)
    
    
ax.errorbar(x=list(range(1, len(average_value_list005) + 1)),y=average_value_list005, yerr=yerr005,
            xerr=None, linestyle='none', ecolor = 'r', elinewidth=0.5)

ax.errorbar(x=list(range(1, len(average_value_list010) + 1)),y=average_value_list010, yerr=yerr010,
            xerr=None, linestyle='none', ecolor = 'y', elinewidth=0.5)

ax.errorbar(x=list(range(1, len(average_value_list025) + 1)),y=average_value_list025, yerr=yerr025,
            xerr=None, linestyle='none', ecolor = 'b', elinewidth=0.5)

ax.errorbar(x=list(range(1, len(average_value_list100) + 1)),y=average_value_list100, yerr=yerr100,
            xerr=None, linestyle='none', ecolor = 'm', elinewidth=0.5)
    
plt.xlabel('Iterations')
plt.ylabel('Groundstate probability')

ax.set_xlim(0)

textstr = '\n'.join((
    r'Red: $\alpha=%.2f$' % 0.05,
    r'Yellow: $\alpha=%.2f$' % 0.10,
    r'Blue: $\alpha=%.2f$' % 0.25,
    r'Magenta: $\alpha=%.2f$' % 1))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax.text(0.02, 0.98, textstr,
        verticalalignment='top', horizontalalignment='left',
        transform=ax.transAxes,
        color='black', fontsize=10, bbox=props)

plt.show()
```
