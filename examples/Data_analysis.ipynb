```python
print(alpha100_samp32_inparams_01_01_df["Ground state probability"])
#varje rad i dessa filer bör vara list eller array men är nu string. behöver ändras innan 
#man kan räkna och plotta med dem. Testa med split(), map, något sånt


#skapa prob_vector
probs=alpha005_samp32_inparams_01_01_df["Ground state probability"] 
floatlists = []
for i in range(1000):
    probabs = probs[i]
    stringlist=(probabs.lstrip("[").rstrip("]").replace(',', '')).split( )
    floatlist= [float(g) for g in np.array(stringlist)]
    floatlists.append(floatlist)
    
    #Vi gör till en lista med floats som går att plotta
```

```python
#Går att plotta men alla i samma färg, ska fixas.
fig, ax = plt.subplots()
for prob_vector in floatlists:
    ax.plot(len(prob_vector), prob_vector[-1], 'y.', linewidth=1)
plt.xlabel('Iterations')
plt.ylabel('Groundstate probability')
plt.show()
```
